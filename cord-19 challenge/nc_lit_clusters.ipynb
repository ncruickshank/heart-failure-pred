```python
# slice 1
meta_df1 = pd.read_csv('metadata1.csv', dtype = {
    'pubmed_id': str,
    'Microsoft Academic Paper ID': str,
    'doi': str
})

# slice 2
meta_df2 = pd.read_csv('metadata2.csv', dtype = {
    'pubmed_id': str,
    'Microsoft Academic Paper ID': str,
    'doi': str
})

# slice 3
meta_df3 = pd.read_csv('metadata3.csv', dtype = {
    'pubmed_id': str,
    'Microsoft Academic Paper ID': str,
    'doi': str
})

# slice 4
meta_df4 = pd.read_csv('metadata4.csv', dtype = {
    'pubmed_id': str,
    'Microsoft Academic Paper ID': str,
    'doi': str
})

# slice 5
meta_df5 = pd.read_csv('metadata5.csv', dtype = {
    'pubmed_id': str,
    'Microsoft Academic Paper ID': str,
    'doi': str
})

# slice 6
meta_df6 = pd.read_csv('metadata6.csv', dtype = {
    'pubmed_id': str,
    'Microsoft Academic Paper ID': str,
    'doi': str
})

# merge
meta = pd.concat([meta_df1, meta_df2, meta_df3,
                  meta_df4, meta_df5, meta_df6])

meta.head()
```
